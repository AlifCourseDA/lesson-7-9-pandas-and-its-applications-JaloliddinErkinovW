```python
path_to_file = "5_data.csv.zip"

retail = pd.read_csv(path_to_file, encoding="ISO-8859-1")
retail_columns = retail.columns
```

```python
#removing duplicates
retail.drop_duplicates(inplace=True)

retail.duplicated().sum()
```

```python
#counting cancelled orders
cancelled = retail[retail["InvoiceNo"].str.startswith("C")]

len(cancelled.InvoiceNo.value_counts())
```

```python
retail = retail[retail["Quantity"] > 0]

len(retail)
```

```python
#users from Germany
germans = retail[retail["Country"] == "Germany"]
#80% percintile
N = np.percentile(germans.InvoiceNo.value_counts(), 80)

germans = (germans
          .groupby("CustomerID", as_index=False)
          .agg({"InvoiceNo" : "count"}))

germans = germans[germans["InvoiceNo"] > N]
germany_top = germans.CustomerID
```

```python
germany_top
```

```python
#top retails in germany using ids from germany_top list
top_retail_germany = retail[retail["CustomerID"].isin(list(germany_top))]

top_retail_germany.head()
```

```python
#filtering by stock code and quantity count
by_stock_code = (top_retail_germany
                .groupby("StockCode")
                .agg({"Quantity" : "count"})
                .sort_values(by="Quantity", ascending=False))

by_stock_code.head()         
```

```python
#another most wanted product
another = retail[retail["StockCode"] == "22326"]

another.Description.head(1)
```

```python
#creating revenue column
retail["Revenue"] = retail.Quantity * retail.UnitPrice

retail.head()
```

```python
#top 5 revenue
top_five = (retail
           .groupby("InvoiceNo", as_index=False)
           .agg({"Revenue" : "sum"})
           .sort_values(by="Revenue", ascending=False))

top_five.head()
```
